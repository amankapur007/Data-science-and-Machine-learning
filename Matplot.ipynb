```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y)
axes2.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])

axes1.plot(x,x*x)
axes2.plot(x,x+20)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.set_xlabel('X')
axes2.set_ylabel('Y')

axes2.set_title('ZOOM')
```

```python
fig, axes=plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,x+20,color="blue", lw=3, ls='--')
axes[1].plot(x,x*x,color="red", lw=3, ls='-')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,x+20,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,x*x,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
